### 243. timedelta
datetime ëª¨ë“ˆì˜ timedeltaë¥¼ ì‚¬ìš©í•´ì„œ ì˜¤ëŠ˜ë¡œë¶€í„° 5ì¼, 4ì¼, 3ì¼, 2ì¼, 1ì¼ ì „ì˜ ë‚ ì§œë¥¼ í™”ë©´ì— ì¶œë ¥í•´ë³´ì„¸ìš”.\
