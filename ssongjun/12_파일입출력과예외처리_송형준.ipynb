## 12. 파일 입출력과 예외처리

파이썬을 이용한 컴퓨터에 저장된 파일을 읽거나 반대로 파일을 쓸 수 있습니다. 프로그램을 작성하다보면 예외가 발생할 수 있는데 이를 잘 처리하는 것이 중요합니다.

### 291 파일쓰기
현재 폴더에 '매수종목1.txt' 파일을 생성하는 코드를 작성하고, 다음과 같이 종목코드를 파일에 써 보세요.

```plain text
005930
005380
035420
```

````python
import os

# 현재 실행 중인 디렉토리: get current workind directory
dir = os.getcwd() 

# 저장할 파일 경로 생성
file_path = os.path.join(dir, "매수종목1.txt")

# 파일 쓰기
with open(file_path, mode="wt", encoding="utf-8") as f: # 파일을 쓰기 위해서는 먼저 open으로 열어야 하죠? 쓰려면 wt 모드로 열어야 해요.
    f.write("005930\n") # .write()을 통해 쓸 수 있어요. 줄바꿈을 하려면 \n 을 써줘야 합니다.
    f.write("005380\n")
    f.write("035420")

print(f"파일이 저장되었습니다: {file_path}")
````

### 293 CSV 파일 쓰기
현재 폴더에 '매수종목.csv' 파일을 생성한 후 다음과 같이 종목코드와 종목명을 파일에 써보세요. 인코딩은 'cp949'를 사용해야 합니다.



````python
import csv # csv 모듈을 통해 간편하게 csv에 쓸 수 있습니다.
import os

# 현재 실행 중인 디렉토리: get current workind directory
dir = os.getcwd() 

# 저장할 파일 경로 생성
file_path = os.path.join(dir, "매수종목.csv")

# 파일 쓰기
f = open(file_path, mode="wt", encoding="cp949", newline='') # 파일에 쓰려면 어떤 모드로 열어야 되더라? 인코딩은 뭐 쓰라고 했죠?
writer = csv.writer(f) # csv 모듈의 writer() 메서드를 통해 손쉽게 csv로 쓸 수 있습니다.
writer.writerow(["종목명", "종목코드", "PER"]) 
writer.writerow(["삼성전자", "005930", 15.59])
writer.writerow(["NAVER", "035420", 55.82])
f.close() # 파일을 다 썼으면 꼭 close로 닫아줘야 합니다!! 
````

### 294 파일 읽기
현재 폴더에 생성한 '매수종목1.txt' 파일을 읽은 후 종목코드를 리스트에 저장해보세요.

```plain text
005930
005380
035420
```

````python
import os

# 현재 실행 중인 디렉토리: get current workind directory
dir = os.getcwd() 

# 저장할 파일 경로 생성
file_path = os.path.join(dir, "매수종목1.txt") # 읽어올 파일의 경로 설정 해봅시다!

# 파일 읽기
f = open(file_path, encoding="utf-8")
lines = f.readlines()   # python list

codes = [] 
for line in lines:
    code = line.strip()  # strip 메서드를 통해 각 줄에 있는 줄바꿈 문자(\n)을 없애줍니다.
    codes.append(code) #리스트에 요소를 추가하려면 append 메서드를 사용하면 됩니다.

print(codes)

f.close()
````

### 295 파일 읽기
현재 폴더에 생성한 '매수종목2.txt' 파일을 읽은 후 종목코드와 종목명을 딕셔너리로 저장해보세요. 종목코드를 key로, 종목명을 value로 저장합니다.

````python
import os

# 현재 실행 중인 디렉토리: get current workind directory
dir = os.getcwd() 

# 저장할 파일 경로 생성
file_path = os.path.join(dir, "매수종목2.txt")

# 파일 읽기
f = open(file_path, encoding="utf-8")
lines = f.readlines()

data = {} # 딕셔너리에 저장해봅시다!
for line in lines:
    line = line.strip()     # strip 메서드를 통해 각 줄에 있는 줄바꿈 문자(\n)을 없애줍니다.
    k, v = line.split()     # split 메서드를 통해 띄어쓰기로 구분되는 값을 분리하여 k, v에 바인딩합시다.
    print('key: {}, value: {}'.format(k, v))             # 확인!
    data[k] = v            # data 딕셔너리에 key는 k, value는 v 값을 저장하려면 어떻게 해야 할까요?

print(data)
f.close()
````

### 296 예외처리
문자열 PER (Price to Earning Ratio) 값을 실수로 변환할 때 에러가 발생합니다. 예외처리를 통해 에러가 발생하는 PER은 0으로 출력하세요.

```python
per = ["10.31", "", "8.00"]

for i in per:
    print(float(i))
```

````python
per = ["10.31", "", "8.00"]

for i in per:
    try:                    # try: 에는 예외가 발생할 수도 있는 코드를 써줍니다.
        print(float(i))
    except:                 # except: 에는 예외 발생 시 실행될 코드를 써줍니다.
        print(0)
````

### 297 예외처리 및 리스트에 저장
문자열로 표현된 PER 값을 실수로 변환한 후 이를 새로운 리스트에 저장해보세요.

```python
per = ["10.31", "", "8.00"]

for i in per:
    print(float(per))
```

````python
per = ["10.31", "", "8.00"]
new_per = []    # 새로운 리스트 new_per에 저장해주세요!

for i in per:
    try:
        v = float(i)    # 실수로 변환합니다.
    except:
        v = 0           # 예외 발생 시 0으로 저장하게 합니다.
    new_per.append(v)   # 리스트에 추가하려면 append 메서드를 써야 했던 거 기억 나죠?

print(new_per)
````

### 298 특정 예외만 처리하기
어떤 값을 0으로 나누면 ZeroDivisionError 에러가 발생합니다. try ~ except로 모든 에러에 대해 예외처리하지 말고 ZeroDivisionError 에러만 예외처리 해보세요.

````python
try:
    b = 3 / 0
except ZeroDivisionError:
    print("0으로 나누면 안돼요")
````

### 299 예외의 메시지 출력하기
다음과 같은 코드 구조를 사용하면 예외 발생 시 에러 메시지를 변수로 바인딩할 수 있습니다.

```python
try:
    실행코드
except 예외 as 변수:
    예외처리코드 
    
```


리스트의 인덱싱에 대해 에러를 출력해보세요.
```python
data = [1, 2, 3]

for i in range(5)
    print(data[i])
```

````python
data = [1, 2, 3]

for i in range(5):
    try:
        print(data[i])
    except IndexError as e:
        print(e)
````

### 300 try, except, else, finally 구조 사용해보기
파이썬 예외처리는 다음과 같은 구조를 가질 수 있습니다.

```python
try:
    실행 코드
except:
    예외가 발생했을 때 수행할 코드
else:
    예외가 발생하지 않았을 때 수행할 코드
finally:
    예외 발생 여부와 상관없이 항상 수행할 코드
```

아래의 코드에 대해서 예외처리를 사용하고 try, except, else, finally에 적당한 코드를 작성해봅시다. else와 finally는 적당한 문구를 print하시면 됩니다.

```python
per = ["10.31", "", "8.00"]

for i in per:
    print(float(per))
```

````python
per = ["10.31", "", "8.00"]

for i in per:
    try:
        print(float(i))
    except:
        print(0)
    else:
        print("clean data")
    finally:
        print("변환 완료")
````
